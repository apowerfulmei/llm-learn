# K均值（K-Means）

---

### K均值算法是什么？

K均值是一种 **聚类（Clustering）** 算法。
它的目标是：
👉 **把数据分成 K 个组（簇）**，让同一组里的数据尽量相似，不同组之间的数据尽量不同。

比如，把一堆水果根据颜色和大小分成 3 类：红苹果一堆、黄香蕉一堆、绿西瓜一堆。

---

### K均值的基本流程

1. **确定 K 值**（想分成几类？比如 3 类）
2. **随机选 K 个数据点**作为**初始中心（质心）**
3. **分配数据点**：
   把每个点分配给**离它最近的中心**。
4. **更新中心**：
   对每个组，重新计算**组内所有点的平均位置**，作为新的中心。
5. **重复**步骤 3 和 4，直到：
   - 中心不怎么变化了，或者
   - 达到最大循环次数

---

### 用一句话总结
> K均值就是不停地**分类 → 重新计算中心 → 再分类**，直到收敛。

---

### 一个小例子

假设有5个点：
`(1,1)，(2,1)，(4,5)，(5,4)，(5,5)`，想分成2类（K=2）。

- 随机选 `(1,1)` 和 `(5,5)` 作为初始中心。
- 计算每个点离哪个中心近，分组。
- 分完组后，算每组的中心，比如 `(1.5, 1)` 和 `(4.67, 4.67)`。
- 用新的中心重新分组。
- 重复，直到中心不变。

最后，就把近的点分到一起了！

---

### 小注意事项

- **K值怎么选？**
  通常用经验，或者方法比如 "肘部法则（Elbow Method）"。
- **初始中心的选择**会影响结果，所以有更好的选法，比如 **K-Means++**。
- K均值只适合**凸形分布**，不太适合像圆环那样复杂的结构。

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. 加载 Iris 数据集
iris = datasets.load_iris()
X = iris.data
y = iris.target  # 真实标签（这里只是用来对比）

# 2. 创建 KMeans 模型，设定分成 3 类
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# 3. 聚类的结果
predicted_labels = kmeans.labels_
print("预测的类别标签:", predicted_labels)

# 4. 把高维数据用 PCA 降到 2 维，方便画图
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 5. 可视化
plt.figure(figsize=(8, 6))

# 用聚类的标签画
for i in range(3):
    plt.scatter(X_pca[predicted_labels == i, 0],
                X_pca[predicted_labels == i, 1],
                label=f'Cluster {i}')

# 画出聚类中心
centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', marker='x', s=200, label='Centers')

plt.legend()
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```
