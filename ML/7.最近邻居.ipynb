## 🧠 一、KNN 是什么？

KNN 是一种**基于实例的学习方法（Instance-based Learning）**，它不对训练数据进行显式的建模，而是保留全部数据，在预测时使用它们来推断新样本的类别或值，因此也被称为**懒惰学习算法（Lazy Learning）**。

---

## 🧩 二、KNN 分类算法原理详解

### 1️⃣ 步骤一：选择K值

- K 是一个正整数，表示要参考的“最近邻”的个数。
- K 值对结果有很大影响：
  - K太小：容易受噪声干扰，模型复杂，容易过拟合。
  - K太大：邻居太多，容易欠拟合，降低模型的区分能力。

👉 一般使用**交叉验证**或经验法则来选择K值。

---

### 2️⃣ 步骤二：计算距离

常见距离度量方法：

#### 🔵 欧几里得距离（最常用）：

对于两个样本 $x = (x_1, x_2, ..., x_n)$ 和 $y = (y_1, y_2, ..., y_n)$，欧几里得距离为：

$$
d(x, y) = \sqrt{ \sum_{i=1}^n (x_i - y_i)^2 }
$$

#### 🔵 曼哈顿距离：

$$
d(x, y) = \sum_{i=1}^n |x_i - y_i|
$$

#### 🔵 闵可夫斯基距离（更通用）：

$$
d(x, y) = \left( \sum_{i=1}^n |x_i - y_i|^p \right)^{1/p}
$$

scikit-learn 默认使用欧几里得距离（`p=2`）。

---

### 3️⃣ 步骤三：选出最近的K个点

- 将训练集中所有样本与测试样本计算距离
- 排序后选择距离最小的K个训练样本

---

### 4️⃣ 步骤四：进行“投票”或“平均”

#### ✅ 分类：

- **多数投票法（Majority Voting）**：K个邻居中哪个类别最多，就预测为哪个类别
- 可以加入**权重**：例如按距离加权，离得近的邻居投票权重更大（`weights='distance'`）

#### 📏 回归：

- 预测值为K个邻居目标值的**平均值**（或加权平均）

---

## 🧬 三、KNN 的一些关键点

### 🔧 1. 特征归一化（标准化）

因为 KNN 依赖距离，如果特征单位差别太大，会导致某些特征主导距离计算。例如“身高（cm）”和“收入（万元）”。

常用方法：
- **标准化（Z-Score）**：让特征均值为0，标准差为1
- **归一化（Min-Max）**：把特征压缩到[0, 1]


---

### 🚀 2. 适用场景和优缺点

#### ✅ 优点：
- 简单易实现
- 无需训练过程，适合快速建模
- 对多分类问题效果好

#### ❌ 缺点：
- 预测慢（每次都要全量计算距离）
- 对维度敏感（“维度灾难”）
- 对噪声和离群点敏感

---

## 📌 四、KNN小总结

| 特性   | 内容             |
|------|----------------|
| 类型   | 监督学习，懒惰学习      |
| 用途   | 分类、回归          |
| 关键参数 | K值，距离函数，权重方式   |
| 要点   | 特征归一化，K选择，计算效率 |

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. 加载数据
iris = load_iris()
X = iris.data       # 特征：花萼长宽、花瓣长宽
y = iris.target     # 标签：三种鸢尾花类别

# 2. 划分训练集与测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. 创建KNN模型，设K=3
knn = KNeighborsClassifier(n_neighbors=3)

# 4. 训练模型
knn.fit(X_train, y_train)

# 5. 预测测试集
y_pred = knn.predict(X_test)

# 6. 评估准确率
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN分类准确率: {accuracy:.2f}")

for i in range(5):
    print(f"样本{i} - 真实标签: {y_test[i]}，预测标签: {y_pred[i]}")
````
