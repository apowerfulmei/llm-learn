# SVM

## 🧠 **SVM 是什么？**

支持向量机（Support Vector Machine）是一种 **监督学习算法**，用于**二分类问题**。
目标是找到一个**最优的分割超平面**，将两类数据尽可能**间隔最大化**地分开。

---

## 🎯 **SVM 的核心思想**

1. **分割线/面（超平面）**：
   - 在二维中，就是一条直线，把两类点分开。
   - 在三维中是一个平面；高维中是超平面。

2. **最大间隔（Margin）**：
   - 我们不光要找到一条分割线，还要找到 **“两类点距离最近的点”之间的间隔最大**的那条线。
   - SVM 要最大化这个间隔。

3. **支持向量（Support Vectors）**：
   - 距离分割线最近的点（边界上）。
   - 它们决定了分割超平面的位置和方向。
   - **只有这些点影响模型，其余点不会影响超平面**。

---

## ✏️ 数学形式（线性可分 SVM）

假设我们有一组点 $(x_i, y_i)$，其中：

- $x_i \in \mathbb{R}^n$：特征向量
- $y_i \in \{+1, -1\}$：类别标签

我们希望找到一个超平面：

$$
w \cdot x + b = 0
$$

使得：

$$
\begin{cases}
w \cdot x_i + b \geq +1, & \text{if } y_i = +1 \\
w \cdot x_i + b \leq -1, & \text{if } y_i = -1
\end{cases}
$$

或者合并写成：

$$
y_i(w \cdot x_i + b) \geq 1
$$

目标函数是最小化 $ \|w\|^2 $，因为：

$$
\text{Margin} = \frac{2}{\|w\|}
$$

所以我们想**最大化间隔**，就变成了：

### 👉 约束优化问题：

$$
\min_{w, b} \frac{1}{2} \|w\|^2 \quad \text{subject to } y_i(w \cdot x_i + b) \geq 1
$$

---

### 💡 为什么是“支持”向量？

因为如果你移除非支持向量，超平面不变；但**移除一个支持向量，超平面就可能改变**。

---

### 🧮 非线性可分怎么办？

用核函数（Kernel Trick）把原始数据**映射到高维空间**，在那个空间中用线性超平面分开。

常用核函数：

- 线性核：$K(x, x') = x \cdot x'$
- 高斯核（RBF）：$K(x, x') = \exp(-\gamma \|x - x'\|^2)$
- 多项式核：$K(x, x') = (x \cdot x' + c)^d$


### 拉格朗日函数

在SVM中，我们的目标是：

$$
\min \frac{1}{2}\|w\|^2
$$

但是！！！有个「超讨厌的约束」：

$$
y_i(w \cdot x_i + b) \geq 1, \quad \forall i
$$

这句话告诉你：每个点都要在正确的边界之外！
这个约束是个**不等式约束**，没办法直接用普通微分去优化。

---

### 这时候，拉格朗日登场！

- 拉格朗日函数 $\mathcal{L}(w,b,\alpha)$ 把约束融进目标。
- 具体做法：每个约束配一个「拉格朗日乘子」$\alpha_i$，然后做加权求和。

公式长这样：

$$
\mathcal{L}(w, b, \alpha) = \frac{1}{2}\|w\|^2 - \sum_i \alpha_i \left( y_i(w \cdot x_i + b) - 1 \right)
$$

- $\frac{1}{2}\|w\|^2$：原始目标
- $\alpha_i(y_i(...)-1)$：约束的“惩罚项”

---

### 拉格朗日的用处到底是什么？

1. **把约束搞进去，变成无约束优化问题**。
   - 拉格朗日的魔法：它的极值，恰好在满足约束的地方！

2. **引入了$\alpha_i$，让SVM可以转化成对偶问题**。
   - 对偶问题更容易解，尤其用核函数的时候。

3. **自动筛选支持向量**。
   - 最后会发现，只有那些$\alpha_i > 0$的点在起作用！它们就是支持向量。

---

### 可视化理解（直觉版）

你可以想象：
- $ \frac{1}{2}\|w\|^2$ 是一座山谷（你想去最底部）。
- $ y_i(w\cdot x+b) - 1 $ 是围墙（约束你不能越界）。
- 拉格朗日乘子 $\alpha_i$ 就像是绑在你身上的橡皮筋——如果你想越界，它会拉住你，把你扯回来。

最后在这座山谷里，你会找到一个既在谷底，又被橡皮筋牵在合法区域的地方。那就是最优解。

---

### 一句话总结：

> **拉格朗日函数在SVM中的用途**：用一堆拉力（乘子）把超平面的参数拉到满足约束的位置，同时尽可能大地开辟间隔。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# 1. 加载数据
iris = datasets.load_iris()
X = iris.data[:, :2]  # 取前两个特征
y = iris.target

# 只选前两个类别（Setosa 和 Versicolor）
mask = y < 2
X = X[mask]
y = y[mask]

# 2. 标准化数据
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 3. 创建 SVM 模型（线性核）
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)

# 4. 获取模型参数（用于画边界）
w = model.coef_[0]
b = model.intercept_[0]
print(f"w = {w}, b = {b:.4f}")

# 5. 可视化决策边界
def plot_svm_decision_boundary(X, y, model):
    # 创建网格点
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    # 预测整个网格
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # 绘图
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')

    # 决策边界 & margin
    x_line = np.linspace(x_min, x_max, 100)
    y_line = -(w[0] * x_line + b) / w[1]
    margin = 1 / np.linalg.norm(w)
    y_upper = y_line + margin
    y_lower = y_line - margin

    plt.plot(x_line, y_line, 'k-', label='Decision Boundary')
    plt.plot(x_line, y_upper, 'k--', linewidth=0.8)
    plt.plot(x_line, y_lower, 'k--', linewidth=0.8)

    # 支持向量
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
                s=100, facecolors='none', edgecolors='k', linewidths=1.5, label='Support Vectors')

    plt.title("SVM with Linear Kernel on Iris Dataset")
    plt.xlabel("Sepal length (standardized)")
    plt.ylabel("Sepal width (standardized)")
    plt.legend()
    plt.grid(True)
    plt.show()

# 6. 画图
plot_svm_decision_boundary(X, y, model)
```
