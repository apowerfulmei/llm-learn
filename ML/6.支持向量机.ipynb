# SVM

支持向量机（Support Vector Machine, 简称 SVM） 是一种二分类模型，属于监督学习中的一种。SVM 的核心思想是：

> 找到一个最优的超平面，把不同类别的数据点尽可能“间隔最大化”地分开。

📌 主要应用：
+ 二分类问题（如垃圾邮件 vs 非垃圾邮件）
+ 图像识别

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# 加载鸢尾花数据集
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# 查看数据集信息
print("特征名称:", feature_names)
print("类别名称:", target_names)
print("样本数量:", X.shape[0])
print("特征数量:", X.shape[1])
```

```python
# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 特征标准化
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 创建SVM分类器
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
# 训练模型
svm_model.fit(X_train, y_train)
```

```python
# 只使用前两个特征进行可视化
X_vis = X_train[:, :2]

# 创建网格来绘制决策边界
x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# 训练仅使用前两个特征的SVM模型用于可视化
svm_vis = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_vis.fit(X_vis, y_train)

# 预测整个网格
Z = svm_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 绘制决策边界和数据点
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=y_train, edgecolors='k')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('SVM / 2 Vectors')
plt.colorbar(ticks=[0, 1, 2], label='Category')
plt.show()
```
